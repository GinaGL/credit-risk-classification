```python
# Import the modules
import numpy as np
import pandas as pd
from pathlib import Path
from pandas.io.common import file_path_to_url
from sklearn.metrics import balanced_accuracy_score, confusion_matrix, classification_report
```

---

## Split the Data into Training and Testing Sets

### Step 1: Read the `lending_data.csv` data from the `Resources` folder into a Pandas DataFrame.

```python
# Read the CSV file from the Resources folder into a Pandas DataFrame
file_path = "lending_data.csv"
df_lending_data = pd.read_csv(file_path)

# Review the DataFrame
display(df_lending_data.head())
display(df_lending_data.tail())
```

### Step 2: Create the labels set (`y`)  from the “loan_status” column, and then create the features (`X`) DataFrame from the remaining columns.

```python
# Separate the data into labels and features

# Separate the y variable, the labels
y = df_lending_data['loan_status']

# Separate the X variable, the features
X = df_lending_data.drop(columns='loan_status')
```

```python
# Review the y variable Series
display(y.head())
display(y.tail())
```

```python
# Review the X variable DataFrame
display(X.head())
display(X.tail())
```

### Step 3: Check the balance of the labels variable (`y`) by using the `value_counts` function.

```python
# Check the balance of our target values
y.value_counts()
```

### Step 4: Split the data into training and testing datasets by using `train_test_split`.

```python
# Import the train_test_learn module
from sklearn.model_selection import train_test_split

# Split the data using train_test_split
# Assign a random_state of 1 to the function
X_train, X_test, y_train, y_test = train_test_split(
    X,
    y,
    random_state = 1
)
```

---

## Create a Logistic Regression Model with the Original Data

###  Step 1: Fit a logistic regression model by using the training data (`X_train` and `y_train`).

```python

# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
LR_model = LogisticRegression(random_state = 1)

# Fit the model using training data
LR_model.fit(X_train, y_train)
```

### Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test`) and the fitted model.

```python

# Make a prediction using the testing data
LR_predictions = LR_model.predict(X_test)
```

### Step 3: Evaluate the model’s performance by doing the following:

* Calculate the accuracy score of the model.

* Generate a confusion matrix.

* Print the classification report.

```python
# Print the balanced_accuracy score of the model
balanced_accuracy_score(y_test, LR_predictions)
```

```python
# Generate a confusion matrix for the model


cm_imbalanced = confusion_matrix(y_test, LR_predictions)
cm_imbalanced_df = pd.DataFrame(cm_imbalanced,
                                index = ['Actual Healthy Loans (low-risk)',
                                'Actual Non-Healthy Loans (high-risk)'],
                                columns = ['Predicted Healthy Loans (low-risk)', 'Predicted Non-Healthy Loans (high-risk)']
                              )
cm_imbalanced_df
```

```python
from sklearn.metrics import classification_report

# Print the classification report
print(classification_report(y_test, LR_predictions))
```

### Step 4: Answer the following question.

**Question:** How well does the logistic regression model predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:**The model achieves a high balanced accuracy score of 95%, indicating its overall performance. However, it's important to interpret this score with caution due to the imbalanced nature of the data. The dataset contains a significantly larger number of healthy loans (low-risk) compared to non-healthy loans (high-risk). As a result, the model tends to predict loan statuses as healthy more accurately than non-healthy.

The imbalanced classification report shows that the model achieves 100% accuracy in predicting healthy loans (class 0). However, its accuracy in predicting non-healthy loans (class 1) is 85%. This difference in accuracy suggests that the model is better at identifying healthy loans than non-healthy loans. It's crucial to consider this data imbalance when evaluating the model's performance and exercise caution when applying it to real-world scenarios.

## Predict a Logistic Regression Model with Resampled Training Data

### Step 1: Use the `RandomOverSampler` module from the imbalanced-learn library to resample the data. Be sure to confirm that the labels have an equal number of data points.

```python
# Import the RandomOverSampler module form imbalanced-learn
from imblearn.over_sampling import RandomOverSampler

# Instantiate the random oversampler model
# # Assign a random_state parameter of 1 to the model
ROS_model = RandomOverSampler(random_state = 1)

# Fit the original training data to the random_oversampler model
X_oversampled, y_oversampled = ROS_model.fit_resample(X_train, y_train)
```

```python
# Count the distinct values of the resampled labels data
y_oversampled.value_counts()
```

### Step 2: Use the `LogisticRegression` classifier and the resampled data to fit the model and make predictions.

```python
# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
LR_oversampled_model = LogisticRegression(random_state = 1)

# Fit the model using the resampled training data
LR_oversampled_model.fit(X_oversampled, y_oversampled)

# Make a prediction using the testing data
LR_oversampled_pred = LR_oversampled_model.predict(X_test)
```

### Step 3: Evaluate the model’s performance by doing the following:

* Calculate the accuracy score of the model.

* Generate a confusion matrix.

* Print the classification report.

```python
# Print the balanced_accuracy score of the model
balanced_accuracy_score(y_test, LR_oversampled_pred)
```

```python
cm_oversampled = confusion_matrix(y_test, LR_oversampled_pred)
cm_oversampled_df = pd.DataFrame(cm_oversampled,
                                index = ['Actual Healthy Loans (low-risk)',
                                'Actual Non-Healthy Loans (high-risk)'],
                                columns = ['Predicted Healthy Loans (low-risk)', 'Predicted Non-Healthy Loans (high-risk)']
                              )
cm_oversampled_df
```

```python
# Print the classification report
print(classification_report(y_test, LR_oversampled_pred))
```

### Step 4: Answer the following question

**Question:** How well does the logistic regression model, fit with oversampled data, predict both the `0` (healthy loan) and `1` (high-risk loan) labels?

**Answer:** The oversampled logistic regression model achieved an impressive accuracy score of 99%, outperforming the model trained on imbalanced data. The improved performance of the oversampled model can be attributed to its ability to effectively identify and correct mistakes, specifically in labeling non-healthy (high-risk) loans as healthy (low-risk). This improvement is evident in the recall score, which increased from 0.91 in the imbalanced model to 0.99 in the oversampled model.

In the lending industry, a higher recall is often preferred. This is because misclassifying healthy loans as non-healthy might lead to some customer dissatisfaction but doesn't have significant financial consequences for the lending company, as they haven't provided any funds to those customers. However, misclassifying non-healthy loans as healthy loans can be quite costly for the lending company, as it results in the loss of funds provided to those customers.

Therefore, the oversampled logistic regression model, with its higher recall and improved ability to identify non-healthy loans, brings valuable benefits to a lending company by reducing the potential financial losses caused by misclassifying risky loans as low-risk loans.
